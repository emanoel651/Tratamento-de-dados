```python
#Importando bibliotecas
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import statistics  as sts
import plotly.express as px
```

```python
#Procurando e importando os dados
destino = 'sample_data/filmes.xlsx'
filme = pd.read_excel(destino)
```

```python
#Mostrando os 100 primeiro dados
display(filme.iloc[:100, :])
```

```python
#Mostrando todos os dados
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
filme
```

```python
#Mudando o nome da coluna 'Genre'
filme = filme.rename(columns={'Genre': 'genero'})
```

```python
#Visualizando novamente
filme.head()
```

```python
#Visualizando os generos dos filmes com apenas horror
filtro = filme.loc[filme['genero'].isin(['Horror'])]
display(filtro)
```

```python
#Visualizando os filme com genero de aventura
horro = filme[filme['genero'].str.contains('Adventure', case=False)]
display(horro)
```

```python
#Visualizando o filme com maior votação
coluna = 'Votes'
votacao = filme['Votes'].max()
display(votacao)
```

```python
#Contando o numero de filmes
display(len(filme))
```

```python
#Mostrando a frequência dos generos
coluna = 'genero'
frequencia = filme[coluna].value_counts().sort_values(ascending=False)
display(frequencia)
```

```python
#Mostrando em gráfico os generos mais assistidos
valores_filtrados = ['Drama', 'Comedy', 'Horror' ]
dados_filtrados = filme[filme[coluna].isin(valores_filtrados)]
frequencia_filme = dados_filtrados[coluna].value_counts()
frequencia_filme.plot(kind='bar')
plt.xlabel(coluna)
plt.ylabel('Frequência')
plt.title(f'Frequência de {coluna}')
plt.show()
```

```python
valores_filtrados = ['Comedy', 'Horror', 'Action,Adventure,Sci-Fi', 'Comedy,Drama,Romance' ]
dados_filtrados = filme[filme[coluna].isin(valores_filtrados)]
frequencia_filme = dados_filtrados[coluna].value_counts()
frequencia_filme.plot(kind='bar')
plt.xlabel(coluna)
plt.ylabel('Frequência')
plt.title(f'Gráfico de Barras - Frequência de Valores em {coluna}')
plt.show()
```

```python
#Mostrando os filmes do gráfico acima
dados_filtrados = filme[filme[coluna].isin(valores_filtrados)]
dados_filtrados.groupby
display(dados_filtrados)
```

```python
#Mudando novamente a coluna 'Votes' para Votação
filme = filme.rename(columns={'Votes': 'Votação'})
```

```python
#Mostrando os filmes mais votados
votos = filme[['Votação', 'Title', 'genero']].sort_values(by='Votação', ascending=False)
display(votos)
```

```python
#Frequência da votaçao
from matplotlib import pyplot as plt
votos['Votação'].plot(kind='hist', bins=20, title='Votação')
plt.gca().spines[['top', 'right',]].set_visible(False)
```
